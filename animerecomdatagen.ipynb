```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import string
```

Used loop for single alphabets one by one because of access blockage on heavy usage

```python
import string
alphabet = list(string.ascii_uppercase)
alphabet.insert(0,'.')
print(alphabet)
```

Formed list of the urls that need to be scraped. 

```python
url_list = []
for a in alphabet:
  url = f'https://myanimelist.net/anime.php?letter={a}'
  for index in range(1,50):
    url_list.append(url)
    url = f'https://myanimelist.net/anime.php?letter={a}'
    url+= f'&show={50*index}'
```

Webscraping the page the find necessary information that are appended to list. 

```python
df_list = []
for url in url_list:
  source = requests.get(url).text
  soup = BeautifulSoup(source, 'lxml')
  articles = soup.findAll('a', class_="hoverinfo_trigger fw-b fl-l")
  for i in articles:
    df_list.append([i['href'].split('/')[4],i.strong.text,i['href']])
```

```python
df = pd.DataFrame(df_list,columns=["anime_id","name","link",])
```

The above data frame is saved as csv format with name anime_df. This is then read for the next steps.

```python
anime_df = pd.read_csv('/content/drive/MyDrive/anime rec/animelist.csv')
anime_df = anime_df.drop(['Unnamed: 0'],axis=1)
anime_df = anime_df.rename(columns={'anime_id': 'uid', 'name': 'title'})
anime_df.head()
```

The anime_df is merged with anime_df2 dataframe found from kaggle to reduce workload of webscraping through the pages. 

```python
anime_df2 = pd.read_csv('/content/drive/MyDrive/anime rec/animes.csv')
anime_df2.drop_duplicates(inplace=True)
anime_df2.head()
```

```python
df = anime_df.merge(anime_df2, on='uid', how='left')
df = df.drop(['title_y','link_y'],axis=1)
df = df.rename(columns={'title_x': 'title', 'link_x': 'link'})
df.head()
```

```python
df = df.drop(['ranked','members','episodes','aired'],axis=1)
```

```python
df_missing = df[df.isna().any(axis=1)]
```

```python
df_missing
```

```python
import time
df_list = []
```

```python
startindex = 0
endindex = 200
for i in range(41):
  for animelink in df_missing['link'][startindex:endindex]:
      genresList = []
      animeSource = requests.get(animelink).text
      soup2 = BeautifulSoup(animeSource, 'lxml')
      synopsis = soup2.find("meta", property="og:description")["content"]
      img_url = soup2.find("meta", property="og:image")["content"]
      score = str(soup2.find("span", class_="score-label").text)
      genres = soup2.findAll("span", itemprop="genre")
      for j in genres:
        genresList.append(j.text)
      df_list.append([animelink,str(genresList),synopsis, score, img_url])
  time.sleep(60)
  startindex+=200
  endindex+=200
```

```python
detail_df = pd.DataFrame(df_list,columns=["link", "genre", "synopsis","score","img_url"])
```
